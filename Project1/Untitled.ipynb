##### No Transformations

```python
x = [1, 4, 3 , 5]
y = [5, 2.75, 3, 2.5]
slope, intercept, r_value1, p_value, std_err = stats.linregress(x,y)
slope, intercept, r_value1, p_value, std_err
```

#####  Exponential

```python
x = [1, 4, 3 , 5]
y = [5, 2.75, 3, 2.5]
y = [np.log(val) for val in y]
slope, intercept, r_value2, p_value, std_err = stats.linregress(x,y)
slope, intercept, r_value2, p_value, std_err
```

##### Power  = 10

```python
x = [1, 4, 3 , 5]
y = [5, 2.75, 3, 2.5]
x = [np.log10(val) for val in x]
y = [np.log10(val) for val in y]
slope, intercept, r_value3, p_value, std_err = stats.linregress(x,y)
slope, intercept, r_value3, p_value, std_err
```

#####  Reciprocal

```python
x = [1, 4, 3 , 5]
y = [5, 2.75, 3, 2.5]
x = [1/(val) for val in x]
slope, intercept, r_value4, p_value, std_err = stats.linregress(x,y)
slope, intercept, r_value4, p_value, std_err
```

##### Hyperbolic

```python
x = [1, 4, 3 , 5]
y = [5, 2.75, 3, 2.5]
x = [1/(val) for val in x]
y = [1/(val) for val in y]
slope, intercept, r_value5, p_value, std_err = stats.linregress(x,y)
slope, intercept, r_value5, p_value, std_err
```
